Condition statement for username and password check

```
username = 'Abhiraj'
password = '@123'

name = input("Enter the usename ")
pswrd = input("Enter the password ")

if name == username :
  if pswrd == password :
    print("Welcome " + username)
  else :
    print("Incorrect Password")
else:
  print("This user is not in the database")
```

Username Password and OTP check

```
username = 'Abhiraj'
password = '@123'
otp = '1112'

name = input("Enter the usename ")
pswrd = input("Enter the password ")
Otp = input("Enter Otp ")

if name == username :
  if pswrd == password :
    if otp == Otp:
      print("Welcome " + username)
    else :
      print("Incorrect OTP")
  else :
    print("Incorrect Password")
else:
  print("This user is not in the database")
```

**Marks and Grading using elif condition**



```
marks = int(input("Enter the mark "))

if marks > 90 :
  print("Grade is A")

elif marks > 80 :
  print("Grade is B")

elif marks >70 :
  print("Grade is C")
  
elif marks > 60 :
  print("Grade is D")

else :
  print ("Failed")
```

**Dictionary**

```
studentMarks = {
    "Ravi" : 90,
    "Raju" : 43,
    "June" : 79,
    "Pooja" : 82,
    "Rahul" : 92
}
i = input("Enter the index ")
studentMarks[i]
```

Students name filter based on the marks

```
studentMarks.keys()
```

```
for key in studentMarks.keys() :
  if studentMarks[key]>60 :
    print(key+ " has passed")
```

```
for (name,marks) in studentMarks.items():
  if marks < 60 :
     print(name + " has passed")   
```
