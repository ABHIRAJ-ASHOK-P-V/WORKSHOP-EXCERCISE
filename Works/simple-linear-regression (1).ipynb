```python
directory = '/kaggle/input/random-linear-regression/'
```

# Extract Data from Csv File

```python
train_data = pd.read_csv (directory + 'train.csv')
test_data = pd.read_csv(directory + 'test.csv')
```

# Derive Information about the Dataset

```python
train_data.describe() 
```

```python
test_data.info()
```

# Removes the Null Value

```python
train_data = train_data.dropna()
train_data.head()
```

```python
train_data.describe()
```

# Extract Input and output from the Train Dataset

```python
X_train = np.array(train_data.iloc[:,0].values)
y_train = np.array(train_data.iloc[:,1].values)
print( 'X_train' , X_train [:5] )
print( 'y_train' , y_train [:5] )
```

# Changing the Shape as LR requires 2D array

```python
X_train = X_train.reshape( -1 ,1 )
print( 'X_train\n' , X_train [:5] )
```

# Extracting Input and Output from Test Dataset

```python
X_test = np.array(test_data.iloc[:,0].values)
y_test = np.array(test_data.iloc[:,1].values)
print( 'X_test' , X_test [:5] )
print( 'y_test' , y_test [:5] )
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit( X_train , y_train )
model.score( X_train , y_train )
```

# Plotting the graph to Check the performance 

```python
import matplotlib.pyplot as plt
plt.scatter( X_test , y_test )
print(X_test.shape)
plt.plot( X_test , model.predict( X_test.reshape( - 1 , 1 ) ) , 'r')
```
